```python
from collections import Counter

def entropy(x):
    
    counter = Counter(x)
    counts = np.array(list(counter.values()))
    total_count = np.sum(counts)
    code_lengths= []
    
    for count in counts:
        code_lengths.append(np.log2(count / total_count))
    
    return -sum(code_lengths) / total_count, total_count
    
```

```python
def uniform_entropy(x):
    return --np.log2(len(set(x)))
```
